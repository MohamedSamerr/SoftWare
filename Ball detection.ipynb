```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
bgr_img = cv.imread('C:\\Users\\moham\\Desktop\\test dataset\\balls\\ball_002.jpg') #please enter the place of image here
rgb_img = cv.cvtColor(bgr_img,cv.COLOR_BGR2RGB)
gray_img = cv.cvtColor(bgr_img,cv.COLOR_BGR2GRAY)
plt.figure(figsize=(10,8))
plt.imshow(rgb_img);
```

```python
blur_img = cv.medianBlur(gray_img,5)
circles = cv.HoughCircles(blur_img,cv.HOUGH_GRADIENT,1.2,120,param1=100,param2=30,minRadius=100,maxRadius=150)
circles = np.uint16(np.around(circles))
```

```python
for i in circles[0,:]:
    cv.circle(rgb_img,(i[0],i[1]),i[2],(0,255,0),4)
    cv.circle(rgb_img,(i[0],i[1]),2,(0,255,0),4)
```

```python
plt.figure(figsize=(10,8))
plt.imshow(rgb_img);
```
